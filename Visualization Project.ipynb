```python
import pandas as pd
import numpy as np

%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
mpl
```

```python
survey=pd.read_csv("Topic_Survey_Assignment.csv", index_col=0)
```

```python
survey.head()
```

```python
survey.sort_values(['Very interested'], ascending=False,axis=0,inplace=True )
survey.head()
```

```python
survey= (survey/2233)
survey.head()
```

```python
surveydf=survey.round(2)
surveydf
```

```python
surveydf.plot(kind='bar', figsize=(20, 8), width=0.8, color=['#5cb85c','#5bc0de','#d9534f']) 
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Choices',fontsize=14) 
plt.ylabel('Percentage',fontsize=14)
plt.title('Percentage of Respondants interest in Data Science Areas', fontsize=16)
plt.legend(fontsize=14)



plt.show()
```

```python
fig= plt.figure() 
ax0 = fig.add_subplot(111)
ax1=surveydf.plot(kind='bar',figsize=(20,8),width = 0.8,color =['#5cb85c','#5bc0de','#d9534f'], ax=ax0)
ax0.set_title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 16)
ax0.set_xlabel('Answers',fontsize= 14)
ax0.set_ylabel('Percent',fontsize= 14)
ax0.set_yticks([])
ax0.spines['right'].set_visible(False)
ax0.spines['top'].set_visible(False)
ax0.spines['left'].set_visible(False)
for p in ax1.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax1.annotate('{:.000%}'.format(height), (x, y + height + 0.01))
plt.show()
```
